```python
# import libraries
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from keras.datasets import mnist
import pandas as pd
```

```python
#importing data
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
```

```python
# target was strength in lab 3  
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength']
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from keras.layers import Flatten , Input

num_trials = 50 #to repeat steps 1-3 50 times
mse_list = []


#Part B : normalize the data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
target_norm=(target - target.mean()) / target.std()

for _ in range(num_trials):
    
    # Step 1: Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target_norm, test_size=0.3, random_state=42)    


    n_cols = X_train.shape[1]
    
    #Model
    model = Sequential()
#     model.add(input_layer)
    model.add(Dense(100, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    
    #part D 2 more layers
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    
    model.add(Dense(1))
    # Compile
    model.compile(optimizer='adam', loss='mean_squared_error')
    #Step 2 :Train using 50 epochs
    model.fit(X_train, y_train, epochs=50, verbose=2)
    # Step 3: Evaluate the model
    y_hat = model.predict(X_test)
    mse = mean_squared_error(y_test, y_hat)
    mse_list.append(mse)
    
```

```python
#Step 5: Report the mean and standard deviation of the mean squared errors
mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

print("Mean of Mean Squared Errors:", mean_mse)
print("Standard Deviation of Mean Squared Errors:", std_mse)
```
